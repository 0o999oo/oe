```python
# Step 1: Gnuplotのインストール
!sudo apt-get install -y gnuplot

# Step 2: gnuplot コマンドを Python コードで実行し、プロットを表示
import subprocess
from IPython.display import Image

# `coba.0.e.ras` ファイルが既に存在するか、ファイルパスを正確に指定してください
# 'output.png' にグラフを保存します
gnuplot_command = """
set terminal png size 800,600;
set output 'output.png';
set xlabel 'Time [s]';
plot [1:2] 'coba.0.e.ras' with dots lc rgb 'black';
"""

# gnuplot コマンドを subprocess を使って実行
with open("temp_script.gp", "w") as f:
    f.write(gnuplot_command)

# 実行
subprocess.run(["gnuplot", "temp_script.gp"])

# 結果の画像を表示
Image(filename="output.png")
```
